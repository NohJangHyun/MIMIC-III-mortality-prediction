```python
import pandas as pd
```

```python
df = pd.read_csv("~/project/MIMIC-III/Data/FINAL/FINAL_Demo+LAB/FINAL_Demo_Lab(1%).csv")

FiO2 = pd.read_csv("~/project/MIMIC-III/Data/OI&OSI/FiO2(all).csv")
MAP = pd.read_csv("~/project/MIMIC-III/Data/OI&OSI/MAP(all).csv")
PaO2 = pd.read_csv("~/project/MIMIC-III/Data/OI&OSI/PaO2(all).csv")
SpO2 = pd.read_csv("~/project/MIMIC-III/Data/OI&OSI/SpO2(all).csv")
```

```python
df = pd.merge(df, FiO2, on=['SUBJECT_ID', 'HADM_ID'], how='left')
df = pd.merge(df, MAP, on=['SUBJECT_ID', 'HADM_ID'], how='left')
df = pd.merge(df, PaO2, on=['SUBJECT_ID', 'HADM_ID'], how='left')
df = pd.merge(df, SpO2, on=['SUBJECT_ID', 'HADM_ID'], how='left')
```

```python
df.to_csv("~/project/MIMIC-III/Data/FINAL/FINAL_Demo+Lab+Chart/FINAL_Demo_Lab_Chart(1%).csv", index=False)
df
```

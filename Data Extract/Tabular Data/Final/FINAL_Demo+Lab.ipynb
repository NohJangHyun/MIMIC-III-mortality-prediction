```python
import pandas as pd
```

```python
df = pd.read_csv("~/project/MIMIC-III/Data/FINAL/Past/Demo/DEMOGRAPHIC_DATA_X.csv")
df1 = pd.read_csv("~/project/MIMIC-III/Data/FINAL/FINAL_LAB/Numeric/Final_lab(100%)_mean.csv")
df2 = pd.read_csv("~/project/MIMIC-III/Data/FINAL/FINAL_LAB/Numeric/Final_lab(100%)_max.csv")
df3 = pd.read_csv("~/project/MIMIC-III/Data/FINAL/FINAL_LAB/Numeric/Final_lab(100%)_min.csv")
ARDS = pd.read_csv("~/project/MIMIC-III/Data/OI&OSI/new_OI&OSI(down).csv")
ARDS = ARDS[["SUBJECT_ID", "HADM_ID", "y"]]
```

```python
# 각 데이터프레임의 칼럼 이름 뒤에 "_mean", "_max", "_min"을 추가
df1.columns = [col + "_mean" for col in df1.columns]
df2.columns = [col + "_max" for col in df2.columns]
df3.columns = [col + "_min" for col in df3.columns]

# 데이터프레임 병합 (기준은 공통 칼럼을 사용하면 됩니다)
df4 = pd.concat([df1, df2, df3], axis=1)

# 겹치는 칼럼 삭제 및 통합
df4.drop(columns=["SUBJECT_ID_max", "HADM_ID_max", "SUBJECT_ID_min", "HADM_ID_min"], inplace=True)
df4.rename(columns={"SUBJECT_ID_mean": "SUBJECT_ID", "HADM_ID_mean": "HADM_ID"}, inplace=True)
df4
```

```python
# SUBJECT_ID와 HADM_ID를 기준으로 데이터프레임 병합
df5 = pd.merge(df, df4, on=['SUBJECT_ID', 'HADM_ID'], how='inner')

# ARDS Label로 변경
df5.drop(columns=['y'], inplace=True)
# df5와 ARDS 데이터프레임을 SUBJECT_ID와 HADM_ID를 기준으로 병합
merged_df = df5.merge(ARDS, on=['SUBJECT_ID', 'HADM_ID'], how='left')
merged_df
```

```python
# 병합된 데이터프레임을 CSV 파일로 저장
merged_df.to_csv("~/project/MIMIC-III/Data/FINAL/FINAL_Demo+LAB/FINAL_Demo_Lab(100%).csv", index=False)
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv("~/project/MIMIC-III/Data/DEMOGRAPHIC_DATA_X.csv")
df2 = pd.read_csv("~/project/MIMIC-III/Data/LABORATORY_VALUE_24h.csv")
df3 = pd.read_csv("~/project/MIMIC-III/Data/LABORATORY_VALUE_24h(30%).csv")
df4 = pd.read_csv("~/project/MIMIC-III/Data/LABORATORY_VALUE_24h(50%).csv")
df5 = pd.read_csv("~/project/MIMIC-III/Data/height_weight_0510.csv")
df6 = pd.read_csv("~/project/MIMIC-III/Data/hr_0510.csv")
df7 = pd.read_csv("~/project/MIMIC-III/Data/temp_0510.csv")
```

```python
df5["weight"] = df5["weight"].fillna(0)
df5["height"] = df5["height"].fillna(0)
df5
```

##### 1. Demo+Lab_24h

```python
df = pd.merge(df1, df2, on=['SUBJECT_ID', 'HADM_ID'], how='right')
df = pd.merge(df, df5, on=['SUBJECT_ID', 'HADM_ID'], how='left')
df = pd.merge(df, df6, on=['SUBJECT_ID', 'HADM_ID'], how='left')
df = pd.merge(df, df7, on=['SUBJECT_ID', 'HADM_ID'], how='left')
df["BMI"] = df["weight"]/(df['height']**2)
df
```

```python
# target_col 칼럼의 인덱스 번호를 구합니다.
Y = df.columns.get_loc('HOSPITAL_EXPIRE_FLAG')

# 칼럼을 임시 리스트에 저장합니다.
cols = df.columns.tolist()

# target_col 칼럼을 마지막으로 이동합니다.
cols = cols[:Y] + cols[Y+1:] + [cols[Y]]

# 변경된 칼럼 리스트를 DataFrame에 적용합니다.
df = df[cols]

x = df["weight"].isnull().sum()
print(x/7584)
x = df["height"].isnull().sum()
print(x/7584)
x = df["hr_min"].isnull().sum()
print(x/7584)
x = df["hr_mean"].isnull().sum()
print(x/7584)
x = df["hr_max"].isnull().sum()
print(x/7584)
x = df["Temperate"].isnull().sum()
print(x/7584)
x = df["BMI"].isnull().sum()
print(x/7584)


df["height"] = df["height"].fillna(0)
df["weight"] = df["weight"].fillna(0)
df["hr_min"] = df["hr_min"].fillna(0)
df["hr_max"] = df["hr_max"].fillna(0)
df["hr_mean"] = df["hr_mean"].fillna(0)
df["Temperate"] = df["Temperate"].fillna(0)
df["BMI"] = df["BMI"].fillna(0)

# df.to_csv("~/project/MIMIC-III/Data/Demo+Lab_24h.csv", index=False)
df
```

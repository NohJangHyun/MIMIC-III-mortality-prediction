```python
df = pd.read_csv("~/project/MIMIC-III/Data/FINAL/Demo/DEMOGRAPHIC_DATA_X.csv")
df1 = pd.read_csv("~/project/MIMIC-III/Data/FINAL/New_Lab/LABORATORY_VALUE_24h(50%)_mean.csv")
df2 = pd.read_csv("~/project/MIMIC-III/Data/FINAL/New_Lab/LABORATORY_VALUE_24h(50%)_max.csv")
df3 = pd.read_csv("~/project/MIMIC-III/Data/FINAL/New_Lab/LABORATORY_VALUE_24h(50%)_min.csv")
ARDS = pd.read_csv("~/project/MIMIC-III/ARDS.csv")
```

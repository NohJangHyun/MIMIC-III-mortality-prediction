##### 6. mechanical ventilation+CPAP, PaO2/FiO2 -> 25,329 ì¤‘ 8,978ëª…

```python
# mechanical ventilation+CPAP, PaO2/FiO2 -> 25,329 ì¤‘ 8,978ëª…
df6 = pd.read_csv("~/project/MIMIC-III/Data/SAPS-II/saps2_mechanical_ventilation.csv")
df7 = pd.read_csv("~/project/MIMIC-III/Data/SAPS-II/saps2_cpap.csv")
df6 = pd.concat([df6, df7], ignore_index=True)
df6.drop('ITEMID', axis=1, inplace=True)
df6 = df6.groupby(['SUBJECT_ID', 'HADM_ID'], as_index=False)['VALUE'].apply(lambda x: 1)
df6.rename(columns={'VALUE': 'VENTILATION_CPAP'}, inplace=True)

df7 = pd.read_csv("~/project/MIMIC-III/Data/SAPS-II/saps2_PaO2.csv")
df8 = pd.read_csv("~/project/MIMIC-III/Data/SAPS-II/saps2_FiO2.csv")
df6 = pd.merge(df6, df7, on=['SUBJECT_ID', 'HADM_ID'], how='outer')
df6 = pd.merge(df6, df8, on=['SUBJECT_ID', 'HADM_ID'], how='outer')

df6.drop('ITEMID_x', axis=1, inplace=True)
df6.drop('ITEMID_y', axis=1, inplace=True)

df6.rename(columns={'VALUE_x':'PaO2', 'VALUE_y':'FiO2'}, inplace=True)
df6 = df6.fillna(0)
df6["VENTILATION_CPAP"] = df6["VENTILATION_CPAP"].astype(int)

df6
```

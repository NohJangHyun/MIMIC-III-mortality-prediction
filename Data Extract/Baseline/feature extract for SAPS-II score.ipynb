#### 2. Data Extract

```python
df = pd.read_csv("~/project/MIMIC-III/Data/SAPS_Feature.csv")
df
```

```python
# 결측치 제거
df.dropna()
# value에 들어간 문자 데이터 제거
df = df[pd.to_numeric(df['VALUENUM'], errors='coerce').notnull()]

# 환자별 첫 24시간 이내의 데이터만 남기기
df['CHARTTIME'] = pd.to_datetime(df['CHARTTIME']) # CHARTTIME 컬럼을 datetime 자료형으로 변환

df = df.groupby(['SUBJECT_ID', 'HADM_ID'], as_index=False).apply(lambda x: x[(x['CHARTTIME'] == x['CHARTTIME'].min()) | ((x['CHARTTIME'] > x['CHARTTIME'].min()) & (x['CHARTTIME'] <= x['CHARTTIME'].min() + pd.Timedelta(hours=24)))])
df = df.reset_index(drop=True)

df
```
